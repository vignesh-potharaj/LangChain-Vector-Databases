```python
from langchain_google_genai import GoogleGenerativeAI

# 1. Initialize the Gemini model using the API key you just set up
llm = GoogleGenerativeAI(
    model="gemini-2.5-flash",
    google_api_key=os.environ.get("GOOGLE_API_KEY",),
    temperature = 0.9
)

# 2. Ask it a test question
try:
    response = llm.invoke("Hey Gemini! Confirming our API connection is active. Give me a 1-sentence greeting.")
    print("🟢 Success! Response from Gemini:")
    print(response)
except Exception as e:
    print("🔴 Something went wrong. Check your API key or installation.")
    print(e)
```
